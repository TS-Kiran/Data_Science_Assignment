## EDA Exploratory Data Analysis

### Customer's EDA

```python
region = df_customers['Region'].value_counts(sort=False)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))
ax = sns.countplot(x='Region', data=df_customers, ax=axes[0])
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', 
                (p.get_x() + p.get_width() / 2., height), 
                ha='center', va='bottom')
axes[0].set_xlabel('Region')
axes[0].set_ylabel('Number of Customers')
axes[0].set_title('Number of Customers per Region')

# Pie chart
axes[1].pie(region, labels=region.index, autopct='%1.1f%%', startangle=140)
axes[1].set_title('Percentage of Customers per Region')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
df_customers['year'] = df_customers['SignupDate'].dt.year
df_customers['month'] = df_customers['SignupDate'].dt.month

# Create subplots
plt.figure(figsize=(12, 6)) 
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot

# Signups by Year (Bar Chart)
ax1 = sns.countplot(x='year', data=df_customers)
plt.xlabel('Year')
plt.ylabel('Number of Signups')
plt.title('Signups by Year')

# Add labels to bars
for p in ax1.patches:
    height = p.get_height()
    ax1.annotate(f'{height}', 
                (p.get_x() + p.get_width() / 2., height), 
                ha='center', va='bottom')

# Signups by Month (Bar Chart)
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
ax2 = sns.countplot(x='month', data=df_customers)
plt.xlabel('Month')
plt.ylabel('Number of Signups')
plt.title('Signups by Month')

# Add labels to bars
for p in ax2.patches:
    height = p.get_height()
    ax2.annotate(f'{height}', 
                (p.get_x() + p.get_width() / 2., height), 
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8)) 

# Create subplots
plt.figure(figsize=(12, 8))

# Signups by Region and Year
plt.subplot(2, 1, 1)  # 2 rows, 1 column, 1st subplot
sns.lineplot(x='year', y=df_customers.index, hue='Region',data=df_customers, 
             legend='full', palette='Set3', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Number of Signups')
plt.title('Signups by Region and Year')
plt.xticks(df_customers['year'].unique())

# Signups by Region and Month
plt.subplot(2, 1, 2)  # 2 rows, 1 column, 2nd subplot
sns.lineplot(x='month', y=df_customers.index, hue='Region', data=df_customers,
             legend='full', palette='Set3', linestyle='-')
plt.xlabel('Month')
plt.ylabel('Number of Signups')
plt.title('Signups by Region')
plt.show()
```

### Transactions EDA

```python
yearly_transactions = df_transactions.groupby(df_transactions['TransactionDate'].dt.year)['TransactionDate'].count()
plt.figure(figsize=(8, 4))
plt.plot(yearly_transactions.index, yearly_transactions.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Transactions')
plt.title('Transactions per Year')

for x, y in zip(yearly_transactions.index, yearly_transactions.values):
    plt.text(x, y, str(y), ha='center', va='bottom') 

plt.xticks(yearly_transactions.index, yearly_transactions.index.astype(int)) 

plt.grid(True)
plt.show()
```

```python
monthly_transactions = df_transactions.groupby(df_transactions['TransactionDate'].dt.month)['TransactionDate'].count()
plt.figure(figsize=(8, 4))
plt.plot(monthly_transactions.index, monthly_transactions.values, marker='o')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.title('Transactions per Month')

# Add labels to each data point
for x, y in zip(monthly_transactions.index, monthly_transactions.values):
    plt.text(x, y, str(y), ha='center', va='bottom')

plt.grid(True)
plt.show()
```

```python
yearly_quantity = df_transactions.groupby(df_transactions["TransactionDate"].dt.year)["Quantity"].sum()
plt.figure(figsize=(8, 4))
plt.plot(yearly_quantity.index, yearly_quantity.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Total Quantity')
plt.title('Yearly Total Quantity')

# Add labels to each data point
for x, y in zip(yearly_quantity.index, yearly_quantity.values):
    plt.text(x, y, str(y), ha='center', va='bottom')

# Find the year with the most quantity
max_quantity_year = yearly_quantity.idxmax()

# Highlight the year with the most quantity
plt.axvline(x=max_quantity_year, color='r', linestyle='--', label=f'Year with Most Quantity: {max_quantity_year}')

# Set x-axis ticks to integer years
plt.xticks(yearly_quantity.index, yearly_quantity.index.astype(int))

# Add legend
plt.legend()

plt.grid(True)
plt.show()
```

```python
monthly_quantity = df_transactions.groupby(df_transactions['TransactionDate'].dt.month)['Quantity'].sum()
plt.figure(figsize=(8, 4))
plt.plot(monthly_quantity.index, monthly_quantity.values, marker='o')
plt.xlabel('Month')
plt.ylabel('Total Quantity')
plt.title('Monthly Total Quantity')

# Add labels to each data point
for x, y in zip(monthly_quantity.index, monthly_quantity.values):
    plt.text(x, y, str(y), ha='center', va='bottom')

# Find the month with the most quantity
max_quantity_month = monthly_quantity.idxmax()

# Highlight the month with the most quantity
plt.axvline(x=max_quantity_month, color='r', linestyle='--', label=f'Month with Most Quantity: {max_quantity_month}')

# Add legend
plt.legend()

plt.grid(True)
plt.show()
```

```python
yearly_price = df_transactions.groupby(df_transactions['TransactionDate'].dt.year)['Price'].sum() 
plt.figure(figsize=(8, 4))
plt.plot(yearly_price.index, yearly_price.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Total Transaction Value')
plt.title('Yearly Total Transaction Value')

# Add labels to each data point
for x, y in zip(yearly_price.index, yearly_price.values):
    plt.text(x, y, f'{y:.2f}', ha='center', va='bottom')  # Format with 2 decimal places

# Find the year with the most price transactions
max_price_year = yearly_price.idxmax()

# Highlight the year with the most price transactions
plt.axvline(x=max_price_year, color='r', linestyle='--', 
           label=f'Year with Most Price Transactions: {max_price_year}')

# Set x-axis ticks to integer years
plt.xticks(yearly_price.index, yearly_price.index.astype(int))

# Add legend
plt.legend()
```
